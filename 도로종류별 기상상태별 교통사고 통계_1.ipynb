# 도로 종류, 기상 상태 교통사고 통계 분석

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False #한글 폰트

df = pd.read_csv('도로종류별 기상상태별 교통사고 통계.csv', encoding='cp949')#데이터불러오기
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.figure(figsize=(15,10))
plt.bar(df['기상상태'], df['사고건수'], color=plt.cm.tab20.colors)
plt.title('기상상태 별 사고 건수')
plt.xlabel('기상상태')
plt.ylabel('사고건수')
plt.show()
```

```python
labels = df["기상상태"]
x = np.arange(len(labels))
width = 0.2

fig, ax = plt.subplots(figsize=(14, 10))
rects1 = ax.bar(x - 2*width, df["사고건수"], width, label='사고건수')
rects2 = ax.bar(x - width, df["사망자수"], width, label='사망자수')
rects3 = ax.bar(x, df["중상자수"], width, label='중상자수')
rects4 = ax.bar(x + width, df["경상자수"], width, label='경상자수')
rects5 = ax.bar(x + 2*width, df["부상신고자수"], width, label='부상신고자수')

ax.set_xlabel('기상상태')
ax.set_title('기상상태별 교통사고 통계')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()
plt.show()
```

```python
df.set_index('기상상태')[['사고건수', '사망자수', '중상자수', '경상자수', '부상신고자수']].plot(kind='bar', stacked=True, color=plt.cm.tab20.colors,figsize=(12, 8))
plt.title('기상상태별 교통사고 통계')
plt.xlabel('기상상태')
plt.ylabel('건수')
plt.tight_layout()
plt.show()
```
