# 도로 종류, 기상 상태 교통사고 통계 분석

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False #한글 폰트

df = pd.read_csv('도로종류별 기상상태별 교통사고 통계.csv', encoding='cp949')#데이터불러오기
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.bar(df['기상상태'], df['사고건수'], color=plt.cm.tab20.colors)
plt.title('기상상태 별 사고 건수')
plt.xlabel('기상상태')
plt.ylabel('사고건수')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.catplot(data=df, kind="bar", x="기상상태", y="사고건수", hue="도로종류", height=6, aspect=1.5)
plt.title('도로 종류와 기상상태에 따른 사고건수')
plt.xlabel('기상상태')
plt.ylabel('사고건수')
plt.tight_layout()
plt.show()
```

```python
road_types = df["도로종류"].unique()

for road_type in road_types:
    subset = df[df["도로종류"] == road_type]
    subset.set_index('기상상태')[['사고건수', '사망자수', '중상자수', '경상자수', '부상신고자수']].plot(
        kind='bar', 
        stacked=True, 
        color=plt.cm.tab20.colors, 
        figsize=(12, 8)
    )
    
    plt.title(f'{road_type} - 기상상태별 교통사고 통계')
    plt.xlabel('기상상태')
    plt.ylabel('건수')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
